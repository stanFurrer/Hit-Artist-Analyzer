```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
DATA_PATH = '../data/spotify/'
mj_df = pd.read_csv(DATA_PATH + 'michael_jackson_tracks.csv')
beatles_df = pd.read_csv(DATA_PATH + 'the_beatles_tracks.csv')
rihanna_df = pd.read_csv(DATA_PATH + 'rihanna_tracks.csv')
eminem_df = pd.read_csv(DATA_PATH + 'eminem_tracks.csv')
sinatra_df = pd.read_csv(DATA_PATH + 'frank_sinatra_tracks.csv')
all_df = sinatra_df.append([beatles_df, mj_df, eminem_df, rihanna_df], ignore_index = True)
```

```python
all_df.to_csv('../data/full_spotify_df.csv')
```

```python
features = all_df.drop('name_track', axis=1).columns
for feature in all_df.drop('name_track', axis=1):
    all_df.hist(feature)
```

```python
correlation = all_df.drop('name_track', axis=1).corr()
fig, ax = plt.subplots(figsize=(20,20))
# Using matshow here just because it sets the ticks up nicely. imshow is faster.
ax.matshow(correlation, cmap='seismic')

for (i, j), z in np.ndenumerate(correlation):
    ax.text(j, i, '{:0.1f}'.format(z), ha='center', va='center')
    
cols = all_df.drop('name_track', axis=1).columns

plt.yticks(np.arange(0, len(cols), 1), cols)
plt.xticks(np.arange(0, len(cols), 1), cols);

plt.show()
```
